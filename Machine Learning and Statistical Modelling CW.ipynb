# Cost Function

```python
def compute_cost(X, y, weights):
    m = len(y)
    predictions = X.dot(weights)
    cost = (1/(2*m)) * np.sum((predictions - y)**2)
    return cost
```

# Gradient Descent Function

```python
def gradient_descent(X, y, weights, learning_rate, iterations):
    m = len(y)
    cost_history = []

    for i in range(iterations):
        predictions = X.dot(weights)
        errors = predictions - y
        gradient = (1/m) * X.T.dot(errors)
        weights -= learning_rate * gradient

        cost = compute_cost(X, y, weights)
        cost_history.append(cost)

    return weights, cost_history
```

```python
# Add bias term (column of ones)
X_train_b = np.c_[np.ones((X_train_scaled.shape[0], 1)), X_train_scaled]
X_test_b  = np.c_[np.ones((X_test_scaled.shape[0], 1)), X_test_scaled]

# Convert y to 2D column vector
y_train_b = y_train.reshape(-1, 1)
y_test_b  = y_test.reshape(-1, 1)
```

```python
weights = np.zeros((X_train_b.shape[1], 1))
```

```python
learning_rate = 0.01
iterations = 400

weights, cost_history = gradient_descent(
    X_train_b, y_train_b, weights, learning_rate, iterations
)
```

```python
plt.plot(cost_history)
plt.title("Cost Function Over Iterations")
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.grid(True)
plt.show()
```

# Evaluating Model Performance

```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

def r2(y_true, y_pred):
    return r2_score(y_true, y_pred)

def metrics_row(name, y_tr_true, y_tr_pred, y_ts_true, y_ts_pred):
    tr_rmse = rmse(y_tr_true, y_tr_pred)
    tr_r2   = r2(y_tr_true, y_tr_pred)
    te_rmse = rmse(y_ts_true, y_ts_pred)
    te_r2   = r2(y_ts_true, y_ts_pred)
    return {"Model": name, "Train RMSE": tr_rmse, "Train R2": tr_r2,
            "Test RMSE": te_rmse, "Test R2": te_r2}

rows = []
# Gradient Descent
rows.append(metrics_row("GradientDescent",
                        y_train, y_train_pred,
                        y_test,  y_test_pred))

# Ridge
rows.append(metrics_row("Ridge",
                        y_train, y_train_pred_ridge,
                        y_test,  y_test_pred_ridge))

# Lasso
rows.append(metrics_row("Lasso",
                        y_train, y_train_pred_lasso,
                        y_test,  y_test_pred_lasso))

# Random Forest
rows.append(metrics_row("RandomForest",
                        y_train, y_train_pred_rf,
                        y_test,  y_test_pred_rf))

# Gradient Boosting
rows.append(metrics_row("GBR",
                        y_train, y_train_pred_gbr,
                        y_test,  y_test_pred_gbr))

# SVR
rows.append(metrics_row("SVR",
                        y_train, y_train_pred_svr,
                        y_test,  y_test_pred_svr))

results_df = pd.DataFrame(rows)
print(results_df)

# Plot: Test RMSE by Model
plt.figure(figsize=(8,4))
sns.barplot(data=results_df, x='Model', y='Test RMSE')
plt.xticks(rotation=45)
plt.title('Test RMSE by Model')
plt.tight_layout()
plt.show()

# Plot: Test R^2 by Model
plt.figure(figsize=(8,4))
sns.barplot(data=results_df, x='Model', y='Test R2')
plt.xticks(rotation=45)
plt.title('Test R^2 by Model')
plt.tight_layout()
plt.show()

# Best model (by lowest Test RMSE)
best = results_df.loc[results_df['Test RMSE'].idxmin()]
print("Best model on test set:", best['Model'])
print("Best Test RMSE:", best['Test RMSE'])
print("Best Test R^2:", best['Test R2'])

# Optional: Actual vs Predicted for the best model (if you want)
# e.g., if best.Model == 'GBR', plot y_test vs y_test_pred_gbr, etc.
```
